```python
import numpy as np
import itertools
from tqdm import tqdm
```

```python
#total number of genes significant in at least one timepoint
N = 2668
#values taken directly from dum of genes with significant average score at each timepoint in chisquare notebook eg 
#eg value calculated in sum(newdf["Day0"]>-np.log10(0.05)) in banach2/wes/Locat/notebooks/figures/Fig3_ESC_Timecourse/ESC_chisquare_analysis-version2_3-cauchy_locat01_repro.ipynb
#T1 T2 T3 T4
n = np.array([432, 515, 1323, 1429])

p1 = n/N
p0 = 1 - p1
```

```python
p1
```

```python
d = []
m = []

for irep in tqdm(range(10000)):
    x = (np.random.uniform(size=(N, 4))<p1[None,:]).astype(float)
    labs, cs = np.unique(x, return_counts=True, axis=0)
    cs = cs/N    
    loc = []
    for lab, c in zip(labs, cs):
        e = np.prod(lab*p1 + (1-lab)*p0)
        loc.append(c-e)
        m.append(c-e)
    d.append(np.max(loc))
```

```python
np.mean(d) 
```

```python
np.std(d)
```

```python
np.mean(m)
```

```python
np.std(m)
```
